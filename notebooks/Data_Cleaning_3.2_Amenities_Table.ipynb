## Data Cleaning Step 3.2
#### Goals for this step
- Create an individual data frame for amenities as columns, 1 indicating True and 0 indicating False

```python
df = pd.read_pickle("C:/Users/Admin/Documents/ironhack/AirBnB_data/airbnb_listings_countries.pkl")
```

```python
df.head()
```

```python
amenities = df["Amenities"].str.split(";")
```

```python
amenities_total = []

for amenity in amenities:
    for i in range(len(amenity)):
        amenities_total.append(amenity[i])
        
amenities_total = list(set(amenities_total))
amenities_total.sort()
amenities_total
```

There are repeated amenities that have different upper and lower cases, we will transform all amenities to lower cases in order to get unique values.

```python
amenities_lower = []

for i in amenities:
    amenities_lower.append([x.lower() for x in i])
```

```python
amenities_count = []

for amenity in amenities_lower:
    for i in range(len(amenity)):
        amenities_count.append(amenity[i])
```

```python
amenities_df = pd.DataFrame(amenities_count)[0].value_counts()
```

```python
amenities_df
```

We will drop unknown amenities:
- translation missing: en.hosting_amenity_49
- translation missing: en.hosting_amenity_50
- other
- other pet(s)

```python
amenities_df = amenities_df.drop(index=["translation missing: en.hosting_amenity_49"
                                        , "translation missing: en.hosting_amenity_50"
                                       , "other"
                                       , "other pet(s)"])
```

We will only use the Top 100 amenities

```python
amenities_df = amenities_df.head(100)
```

```python
amenities_df.index
```

```python
amenities_top = amenities_df.reset_index().rename(columns={"index":"amenity"})["amenity"]
```

Creating blank amenities data frame

```python
amenities_df = pd.DataFrame(columns=(amenities_top), index=range(df.shape[0]))
```

```python
amenities_df
```

```python
rows = range(df.shape[0])

def amenities_func(row, df, amenities_lower):
    for i in amenities_lower[row]:
        if i in df.columns:
            df.loc[row, i] = 1
        else:
            continue
    return df
```
