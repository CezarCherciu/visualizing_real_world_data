```python
# count null values
```

```python
''' 
- SQL -
USE airbnb;
CREATE TABLE listings_withOUT_rating
SELECT `Listing ID`, `Review Scores Rating`, `Last Review Date`
FROM airbnb_ratings_new
WHERE `Last Review Date` IS NULL;
'''
# 634,891 rows
```

```python
# remove empty rows, remove rows with null and 0 review scores
'''
USE airbnb;
CREATE TABLE listings_with_rating
SELECT *
FROM airbnb_ratings_new
WHERE `Last Review Date` IS NOT NULL;
'''
# 412,068 rows
```

```python
pd.read_sql("SELECT * FROM listings_with_rating", conn)
```

```python
# create subset of listings with ratings  OUTSIDE of the United States
```
